```python
balance = 10000
pin = 1123

def atm_simulator():

    print(f"**********Welcome To The ATM Machine Simulator**********")
    print(f"")

    pin_check=int(input("Enter Your Pin: "))
    if pin_check==pin:
        print(f"")
        print(f"Select The Below Options")
        print(f"")
        print(f"1.) Check Available Balance")
        print(f"2.) Withdraw Amount")
        print(f"3.) Deposit Amount")
        print(f"4.) exit")
        print(f"")
        option=int(input(""))
        print(f"")
        if option==1:
            print(f"Your Selected Option is: 1.) Check Available Balance ")
            print(f"")
            print(f"Your Available Balance is: ₹{balance}")
            print(f"")
            print(f"***Thanks For Using ATM Machine Simulator***")
        elif option==2:
            print(f"Your Selected Option is: 2.) Withdraw Amount ")
            print(f"")
            Withdraw_amount=int(input("Enter The Amount: ₹"))
            if Withdraw_amount>10000:
                print(f"")
                print(f"Insufficient Balance!!")
                print(f"")
                print(f"***Thanks For Using ATM Machine Simulator***")
            elif Withdraw_amount<=10000:
                new_balance=balance-Withdraw_amount
                print(f"")
                print(f"Available Balance is: ₹{new_balance}")
                print(f"")
                print(f"***Thanks For Using ATM Machine Simulator***")

        elif option==3:
            print(f"Your Selected Option is: 3.) Deposit Amount ")
            print(f"")
            Deposit_amount=int(input("Enter The Amount: ₹"))
            if Deposit_amount<1:
                print(f"You Have Entered null value")
                print(f"")
                print(f"***Thanks For Using ATM Machine Simulator***")
                
            elif Deposit_amount>0:
                new_balance1=balance+Deposit_amount
                print(f"")
                print(f"Your Available Balance is: ₹{new_balance1}")
                print(f"")
                print(f"***Thanks For Using ATM Machine Simulator***")
        elif option==4:
            print(f"Your Selected Option is: 4.) exit ")
            print(f"")
            print(f"***Thanks For Using ATM Machine Simulator***")
    else:
        print(f"")
        print(f"You Have Entered Incorrect Pin Transaction Cancelled ")
        print(f"")
        print(f"***Thanks For Using ATM Machine Simulator***")
        
atm_simulator()
```
